```python
import numpy as np
import matplotlib.pyplot as plt
import random
from math import e
import math

#constantes
dim=2 #dimension de la particula
n=100 #cantidad de particulas del enjambre
t=1 #tiempo
w=0.70 #coeficiente de inercia
c1=0.20 #coeficiente de aceleracion
c2=0.60 #coeficiente de aceleracion
ite=100 #numero de iteraciones
print(ite,"iteraciones")
lbp=[] #local best position
lbf=[] #local best fitness
gbp=0 #global best position
gbf=0 #global best fitness

def funcionEASOM(x1,x2): 
    cuenta=-math.cos(x1)*math.cos(x2)*math.exp(-((np.power(x1- math.pi,2) + np.power(x2- math.pi,2))))
    return cuenta
    
def genera_particula_velocidad(): #genera vector entre 0 y 1 - r1 y r2
    vector_ini=np.zeros([n,dim])
    for i in range(0,n):
        vector_ini[i,0]=random.random()
        vector_ini[i,1]=random.random()
    return vector_ini

def init_v_new(r2): #esta funcion pasa de r2 a v
    v=np.zeros([n,dim])
    for i in range(0,n):
        v[i,0]=r2[i,0]-0.5
        v[i,1]=r2[i,1]-0.5
    return v

def init_cp_new(r1): #esta funcion pasa de r1 a cp
    cp=np.zeros([n,dim])
    for i in range(0,n):
        cp[i,0]=10*(r1[i,0]-0.5)
        cp[i,1]=10*(r1[i,1]-0.5)
    return cp

def fitness_new(posiciones): #calcula el vector cf de f(posiciones)
    cf=[]
    for i in range(int(n)): #length de posiciones tambien
        cf.append(funcionEASOM(posiciones[i,0],posiciones[i,1]))
    return cf

def update_posicion_new(cp,v): # actualiza el vector poblacion posicion(i+1) = posicion(i)+velocidad(i+1)... v es despues de update_velocidad
    posiciones_actualizada=np.zeros([n,dim])
    for i in range(int(n)):
        posiciones_actualizada[i,0]=cp[i,0]+v[i,0]
        posiciones_actualizada[i,1]=cp[i,1]+v[i,1]
    return posiciones_actualizada
    
def update_velocidad_new(v,lbp,gbp,r1,r2,cp):
    for i in range(1):#aca tuve problemas con el indice, aca extraigo los valores del vector gbp
        gbp1=gbp[0]
        gbp2=gbp[1]
    velocidades_actualizada=np.zeros([n,dim])
    for i in range(int(n)):
        velocidades_actualizada[i,0]=(w*v[i,0])+((c1*r1[i,0])*(lbp[i,0]-cp[i,0]))+(c2*r2[i,0]*(gbp1-cp[i,0]))
        velocidades_actualizada[i,1]=(w*v[i,1])+((c1*r1[i,1])*(lbp[i,1]-cp[i,1]))+(c2*r2[i,1]*(gbp2-cp[i,1]))
    return velocidades_actualizada

def mejor_fitness(cf):
    return (min(cf))

def posicion_mejor_fitness(cf,posiciones):
    posicion=(cf.index(min(cf)))
    return posiciones[posicion]    

#estado inicial
r3=genera_particula_velocidad()
r4=genera_particula_velocidad()
print("genero r1( t=",t,")",r3)
print("genero r2( t=",t,")",r4)
print("####################")
print("Tiempo=",t)
cpnew=init_cp_new(r3)
vnew=init_v_new(r4)
print("genero cp(",t,")",cpnew)
print("genero v(",t,")",vnew)
#inicializo lbp,lbf,gbp y gbf ->primer iteración
lbp=cpnew
lbf=fitness_new(cpnew)
print("genero cf(",t,")",lbf)
gbf=mejor_fitness(lbf)
gbp=posicion_mejor_fitness(lbf,cpnew)
print("Mejor Posición Local (LBP)=",lbp)
print("Mejor Fitness Local(LBF)=",lbf)
print("Mejor Fitness Global (GBF)=",gbf)
print("Mejor Posicion Global (GBP)=",gbp)
#estado inicial

for i in range(int(ite)):
    #calculo nuevas velocidades
    #calculo nuevas posiciones
    #analizo lbp,lbf,gbp y gbf
    print("####################")
    t=t+1
    print("Tiempo=",t)
    v_actualizada=update_velocidad_new(vnew,lbp,gbp,r3,r4,cpnew) #actualizo velocidades
    print("v(",t,")",v_actualizada)
    cp_actualizada=update_posicion_new(cpnew,v_actualizada) #actualizo posiciones
    print("cp(",t,")",cp_actualizada)
    #aplico cf a cp(i)
    cf_actualizada=fitness_new(cp_actualizada)
    print("cf(",t,")",cf_actualizada)
    #actualizo los valores
    lbp=cp_actualizada
    lbf=fitness_new(cp_actualizada)
    gbf=mejor_fitness(lbf)
    gbp=posicion_mejor_fitness(lbf,cp_actualizada)
    print("Mejor Posición Local (LBP)=",lbp)
    print("Mejor Fitness Local(LBF)=",lbf)
    print("Mejor Fitness Global (GBF)=",gbf)
    print("Mejor Posicion Global (GBP)=",gbp)
    plt.plot(lbp, lbf, 'o', color='green')
print("####################")    
print("Mejor Fitness Global (GBF)=",gbf)
#print("Mejor gbf=",format(gbf, '.6f'))
print("Mejor Posicion Global (GBP)=",gbp)
#plot
plt.title("Función EASOM")
#plt.xlabel("Variable X2")
#plt.ylabel("Variable X1")
plt.plot(math.pi,gbf ,'o', color='red',markersize = 20)
plt.grid(True)
plt.annotate('Mejor Fitness Global (Minimo Global) = '+str(format(gbf, '.6f')), xy =(math.pi,gbf), xytext =(6,-0.6), arrowprops = dict(facecolor ='red',shrink = 0.1),)
```
